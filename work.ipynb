```python
label_df = pd.read_csv(os.path.join(csv_files[0]))
label_df.head()
```

```python
csv = pd.read_csv(csv_files[1])
# csv.iloc[1, 0]
# df_columns
csv 
```

```python
label_inx = 0
for idx, csv in enumerate(csv_files[1:]):
    df = pd.read_csv(csv)
    
    df_columns = df.columns
    # print('===>', len(df_columns))
    start_num = 2 if idx > 0 else 4 
    for num in range(start_num, len(df_columns)+2, 2):
        if num >= len(df_columns):
            break
        # print(f'num : {num} // label : {label_inx}')
        df.at[10, df_columns[num]] = label_df.loc[label_inx, 'diagnosis']
        
        if num+1 >= len(df_columns):
            break
        # print(f'num : {num+1} // label : {label_inx}')
        df.at[10, df_columns[num+1]] = label_df.loc[label_inx, 'diagnosis']    
        label_inx += 1 
        
    df.to_csv(os.path.join('C:/Users/user/Desktop/models/', f'TEST 1-{idx+1}.csv'))
    # print(csv)

label_inx  
```
